## Scraping Structured Data from Tables

We extract color names and corresponding hex codes from an HTML table.

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/HTMLColorCodes.html"
```

```python
data = requests.get(url).text
```

```python
soup = BeautifulSoup(data, "html.parser")
```

```python
table = soup.find('table')
table
```

```python
for row in table.find_all('tr'):  # each row
    cols = row.find_all('td')     # each column

    if len(cols) > 0:  # skip header row
        color_name = cols[2].string
        color_code = cols[3].string

        print("{} ---> {}".format(color_name, color_code))
```

```python
for row in table.find_all('tr'):
    cols = row.find_all('td')

    if len(cols) > 0:
        color_name = cols[2].text.strip()
        color_code = cols[3].text.strip()

        print(f"{color_name} ---> {color_code}")
```

## Real-World Web Scraping Example

In this section, we scrape data from Wikipedia and convert it into a structured dataset.

Steps include:
- locating the correct table
- extracting rows and columns
- building a Pandas DataFrame

```python
url = "https://en.wikipedia.org/wiki/World_population"
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/91.0.4472.124 Safari/537.36"
}
```

```python
data = requests.get(url, headers=headers)
```

```python
soup = BeautifulSoup(data.text, "html.parser")
```

```python
tables = soup.find_all('table')
len(tables)
```

```python
for index, table in enumerate(tables):
    if ("10 most densely populated countries" in str(table)):
        table_index = index

print(table_index)
```

```python
population_data = pd.DataFrame(columns=["Rank", "Country", "Population", "Area", "Density"])
```

```python
for row in tables[table_index].tbody.find_all("tr"):
    col = row.find_all("td")

    if col:
        rank = col[0].text.strip()
        country = col[1].text.strip()
        population = col[2].text.strip()
        area = col[3].text.strip()
        density = col[4].text.strip()

        new_row = pd.DataFrame([{
            "Rank": rank,
            "Country": country,
            "Population": population,
            "Area": area,
            "Density": density
        }])

        population_data = pd.concat([population_data, new_row], ignore_index=True)
```
